 В этом задании требуется определить в чем состоит нарушение принципа и предложить переработку кода для устранения нарушения

Open/Closed Principle (OCP) — Принцип открытости/закрытости: Программные сущности должны быть открыты для расширения, но закрыты для модификации. Это значит, что можно добавлять новую функциональность, не меняя существующий код.

```
class Rectangle(object):

    def __init__(self, width, height):
        self.width = width
        self.height = height

class AreaCalculator(object):

    def __init__(self, shapes):

        assert isinstance(shapes, list), "`shapes` should be of type `list`."
        self.shapes = shapes

    @property
    def total_area(self):
        total = 0
        for shape in self.shapes:
            total += shape.width * shape.height

        return total

def main():
    shapes = [Rectangle(2, 3), Rectangle(1, 6)]
    calculator = AreaCalculator(shapes)
    print("The total area is: ", calculator.total_area)

if __name__ == '__main__':

    main()
```
