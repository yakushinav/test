 Описание входного формата

[ГГГГ-ММ-ДД ЧЧ:ММ:СС] CODE: Message (user_id=5821; amount=1500; error=C023)
- CODE – это вид записи лога одно из трех значений: INFO, ERROR, WARNING;
- Message – пояснительное сообщение набор символов английского алфавита и пробелов длиной от 5 до 30 символов;
- user_id – это идентификатор клиента, целое число от 1000 до 9999 включительно;
- amount – это сумма платежа, целое число от 1 до 1000000 включительно;
- error – это код ошибки, имеющий следующий формат XNNN, X это один из символов F,C,A, NNN это число от 000 до 999 включительно.
На вход подаются строки в следующем формате:
[ГГГГ-ММ-ДД ЧЧ:ММ:СС] CODE: Message (user_id=5821; amount=1500; error=C023)
